```python
# Basic setup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better plots
%matplotlib inline
sns.set(style="whitegrid")
```

# Load the dataset

```python
# Replace with actual path
df = pd.read_csv("C:/Users/98939/Downloads/MASTER_DATASET.csv")

# show info
print("Dataset Shape:", df.shape)
print("\n Missing Values per Column:\n")
print(df.isnull().sum())
```

# Statistical Overview

```python
print("Statistical Summary:\n")
display(df.describe())
```

# Histograms of Key Features

```python
# select key columns
columns_to_plot = [
    "Pressure_In",
    "Temperature_In",
    "Flow_Rate",
    "Pressure_Out",
    "Vibration",
]

df[columns_to_plot].hist(bins=50, figsize=(15, 8), color="skyblue", edgecolor="black")
plt.suptitle("Distribution of Key Variables", fontsize=16)
plt.tight_layout()
plt.show()
```

# Boxplot for Outlier Detection

```python
# باکس‌پلات برای بررسی داده‌های پرت
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[columns_to_plot], palette="coolwarm")
plt.title("Boxplot for Key Variables", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

# Correlation Matrix (Heatmap)

```python
# رسم heatmap برای بررسی همبستگی
plt.figure(figsize=(16, 12))
corr = df.corr(numeric_only=True)  # فقط ستون‌های عددی
sns.heatmap(corr, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numeric Features", fontsize=16)
plt.show()
```
